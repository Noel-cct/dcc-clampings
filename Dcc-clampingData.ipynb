```python
from pandas_ods_reader import read_ods
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)

# Reading data from file, first sheet
data_sheet1 = read_ods("/Users/Noel/Documents/MSc Data Analytics/dccclamping.ods")

data_sheet1.head(5)

# Renaming columns
data_sheet1.columns = data_sheet1.iloc[0]

# Checking renaming has worked
data_sheet1.head(5)

# Assigning variables to be graphed
x_year = ['2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011']
y_clamps = data_sheet1.iloc[1, 1:-2]

y_clamps

# Make barchart of x & y values
fig = plt.figure(figsize=(10, 5))
plt.bar(x_year, y_clamps, width=0.4)
plt.xlabel("Year")
plt.ylabel("Clampings")
plt.title("Clampings per Year")
plt.show()

# Creating new variable for clamp removed data
y_removals = data_sheet1.iloc[3, 1:-2]

y_removals

x_axis = np.arange(len(x_year))

plt.bar(x_axis - 0.2, y_clamps, 0.4, label='Clampings')
plt.bar(x_axis + 0.2, y_removals, 0.4, label='Removals')

plt.xticks(x_axis, x_year)
plt.xlabel("Year")
plt.ylabel("Clampsing/Removals")
plt.title("Clampings/Removals Per Year")
plt.legend()
plt.show()

x_axis = np.arange(len(x_year))

plt.bar(x_axis - 0.2, y_clamps, 0.4, label='Clampings')
plt.bar(x_axis + 0.2, y_removals, 0.4, label='Removals')

plt.xticks(x_axis, x_year)
plt.xlabel("Year")
plt.ylabel("Clampsing/Removals")
plt.title("Clampings/Removals Per Year")
plt.legend()
plt.show()

# Checking dataset data
data_sheet1.head()

# Creating variables for enforcements and relocations variables
y_enforcements = data_sheet1.iloc[4, 1:-2]
y_relocations = data_sheet1.iloc[2, 1:-2]

# Make barchart of x & y values
fig = plt.figure(figsize=(10, 5))
plt.bar(x_year, y_enforcements, width=0.4)
plt.xlabel("Month")
plt.ylabel("Enforcements")
plt.title("Enforcements per month")
plt.show()

# Creating histogram with seaborn library
ax = sns.histplot(y_enforcements, bins=10)
ax.set(xlabel='Enforcements', ylabel='Frequency', title='Enforcements Histogram')

plt.plot(x_year, y_enforcements, label="enforcements")
plt.legend()
plt.show()

plt.plot(x_year, y_clamps, label="Clampings")
plt.plot(x_year, y_enforcements, label="Enforcements")
plt.plot(x_year, y_removals, label="Removals")
plt.plot(x_year, y_relocations, label="Relocations")

plt.xlabel("Year")
plt.ylabel("Events")
plt.title("Events per Year")
plt.legend()
plt.show()

# Reading data from file, 4th sheet
data_sheet4 = read_ods("/Users/Noel/Documents/MSc Data Analytics/dccclamping.ods", 4)

data_sheet4.head(5)

# Removing blank columns and rows from data frame
data_sheet4.columns = data_sheet4.iloc[3]
data_sheet4 = data_sheet4[4:]

# Creating variable for months
x_month = data_sheet4.iloc[0:12, 0]

data_sheet4.head(5)

y_clamps = data_sheet4.iloc[0:12, 1]

y_relocations = data_sheet4.iloc[0:12, 2]
y_removals = data_sheet4.iloc[0:12, 3]

x_axis = np.arange(len(x_month))

plt.bar(x_axis - 0.2, y_clamps, 0.4, label='Clampings')
plt.bar(x_axis + 0.2, y_relocations, 0.4, label='Relocations')
plt.bar(x_axis + 0.6, y_removals, 0.4, label='Removals')

plt.xticks(x_axis, x_month)
plt.xlabel("Year")
plt.ylabel("Clampsing/Removals")
plt.title("Clampings/Removals Per Year")
plt.legend()
plt.show()

x_axis = np.arange(len(x_month))

plt.bar(x_axis - 0.4, y_clamps, 0.4, label='Clampings')
plt.bar(x_axis + 0.4, y_relocations, 0.4, label='Relocations')
plt.bar(x_axis, y_removals, 0.4, label='Removals')

plt.xticks(x_axis, x_month)
plt.xlabel("Year")
plt.ylabel("Clampsing/Removals")
plt.title("Clampings/Removals Per Year")
plt.legend()
plt.show()

data_sheet4 = data_sheet4.iloc[0:12]

data_sheet4.head(14)

data_sheet4.describe()

mean_clamps = data_sheet4["Clamps"].mean()
print(mean_clamps)

x = data_sheet4.iloc[0:12, 1]

mean_clamps = data_sheet4.iloc[0:12, 1].mean()
print(mean_clamps)

median = data_sheet4.iloc[0:12, 1].median()
print(median)

mode = data_sheet4.iloc[0:12, 1].mode()
print(mode)

ax = sns.histplot(data_sheet4.iloc[0:12, 1], bins=10)
ax.set(xlabel='Clampings', ylabel='No. of Months', title='Clampings Histogram')

plt.grid(axis='y', alpha=0.75)

ax = sns.histplot(data_sheet4.iloc[0:12, 1], color='green', bins=10)
ax.set(xlabel='Clampings', ylabel='No. of Months', title='Clampings Histogrm')

plt.figure(figsize=(8, 8))
ax = sns.barplot(x="Month", y="Clamps", data=data_sheet4.iloc[0:12, 0:2])
```
